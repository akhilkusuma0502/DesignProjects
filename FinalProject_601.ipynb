# When retrieving my groupby values, i have found that , the group by doesn't understand the column names that i pass. For this I have to Backtrack the code and my dataframe to check for other names or formats exist in the dataframe column.

```python
Enacted_Budget_DataFrame.columns
```

# How i discovered that there is '\xa0' Latin1 code in the python object.
# \xa0 is actually non-breaking space in Latin1

```python
unformatted_columns=[]
for column in Enacted_Budget_DataFrame.columns:
    unformatted_columns.append(column)
unformatted_columns
```

# I have found many solutions for that in stack overflow, but i preferred string object replace function to replace those characters with empty character which eventually solved my problem.

# I have written the code as a user-defined function just to make sure that there may be possibility of similar problem in other datafile. At this case, code reusability is the best option.

```python
## User-defined Function(UDF)
## This expects a parameter which is a list.
## This returns a list.
def removeUnnecessarySpaces(list_of_columns):
    formatted_columns=[]
    for column in list_of_columns:
        formatted_columns.append(column.replace('\xa0','').replace(' ',''))
    return formatted_columns
```

# I am calling my UDF just to remove the latin1 non breaking space from the dataframe

```python
Enacted_Budget_DataFrame.columns=removeUnnecessarySpaces(Enacted_Budget_DataFrame.columns)
Enacted_Budget_DataFrame.columns
```

# Exactly as above, I have called the same function for my another dataframe

```python
Executive_Budget_DataFrame.columns=removeUnnecessarySpaces(Executive_Budget_DataFrame.columns)
Executive_Budget_DataFrame.columns
```

# 1 Plotting Pie Chart  for how much more the Appropriations are recommended in each Appropriation categories.

# Step 1: Summing up all the Appropriations Available based on Appropriation category from the Enacted DataFrame

```python
get_Enacted_Approproations_Category_wise_budget=Enacted_Budget_DataFrame.groupby('AppropriationCategory')['AppropriationsAvailableFY2017'].sum().to_dict()
get_Enacted_Approproations_Category_wise_budget
```

# Step 2: Summing up all the Appropriations Recommended based on Appropriation category from the Executive DataFrame

```python
get_Executive_Approproations_Category_wise_budget=Executive_Budget_DataFrame.groupby('AppropriationCategory')['AppropriationsRecommended2017-18'].sum().to_dict()
get_Executive_Approproations_Category_wise_budget
```

```python
Appropriation_Category_dataframe=pd.DataFrame([get_Enacted_Approproations_Category_wise_budget,get_Executive_Approproations_Category_wise_budget])
Appropriation_Category_dataframe
```

# The reason why i have converted two dictionaries to dataframe is , it is for a dataframe to get the difference of two records using the period value. 

```python
Appropriation_Category_dataframe=Appropriation_Category_dataframe.astype('int64')
Appropriation_Category_dataframe=Appropriation_Category_dataframe.diff().drop(index=0)
Appropriation_Category_dataframe
```

# I am now extracting the records as list of values,.

```python
Difference=[]
for index, rows in Appropriation_Category_dataframe.iterrows():
    Difference=[rows['Aid To Localities'],rows['Capital Projects'],rows['Debt Service'],rows['State Operations']]
Difference
```

So, i got what i wanted, it is easy for me to visualize the above outcomes using piechart which tells me the what amount of value is greater of the whole.

```python
_=plt.figure(figsize=(30,15))
_=plt.pie(Difference,labels=list(Appropriation_Category_dataframe.columns),autopct='%1.1f%%',textprops={'fontsize': 25})
_=plt.title('Pie Chart Showing Extra Recommended Appropriations based on Category Wise for the Year 2017-2018',fontsize=35)
_=plt.show()
```

# By Observing the Above Pie Chart , Aid to Localities Category has been Recommended with more budget Appropriation in the assessment year 2017-2018 
# Whereas Capital Projects Category with 0 budget appropriations.
