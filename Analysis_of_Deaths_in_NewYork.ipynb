```python
# This function loops over the rows of data frame and checks whether the cell value is not persistant. 
# i.e. either null, Nan or a '.'
def dataClean():                                                   
    for index,row in dframe.iterrows():
        for column in columnNames:
            if dframe.at[index,column]=='.' or dframe.at[index,column]=='NaN':
                dframe.at[index,column]=numpy.nan
```

At any time , i can call this function to replace those inpersistent rows and replace it with NaN.

```python
# The cleaning function is called here
dataClean()
```

```python
# changing the data types of each columns 
# astype is used to change the data type of a column
dframe['Deaths'] = dframe['Deaths'].astype('float64')
dframe['Race Ethnicity'] = dframe['Race Ethnicity'].astype('category')
dframe['Sex'] = dframe['Sex'].astype('category')
dframe['Leading Cause'] = dframe['Leading Cause'].astype('category')
```

```python
#This function will generate filtered data Which is a summation of all unique items.
# It requires three parameters 
# ColumnName1 is the column of which we need to compare.
#ColumnName2 is the column we need to sum if certain condition Satisfied.
#ColumnList is list of unique values generated from the column.
def getFilteredSum(columnName1,columnName2,ColumnList):
    tempList=[]
    for item in ColumnList:
        print(item," : ",dframe[dframe[columnName1]==item][columnName2].sum())  ##This will print all the summation values
        tempList.append(dframe[dframe[columnName1]==item][columnName2].sum())   ##I am appending the sum values to a list so that i can use it to plot.
    return tempList
```

```python
# This logic copies all the unique value of Race Ethnicity to a list
getEthnicityList=dframe['Race Ethnicity'].unique().tolist()
```

```python
# This logic copies all the unique values of year to list
getYearList=dframe['Year'].unique().tolist()
```

```python
# This logic copies all the uniques values in sex column to a list
getGenderList=dframe['Sex'].unique().tolist()
```

```python
# This diplay the unique values of Ethnicity list
getEthnicityList
```

```python
# This diaplays the unique values of Year list
getYearList
```

```python
# this displays the unique values of Sex column
getGenderList
```

```python
# This logic calls the function getFilteredSum
# The input aprameters are Column names and ethnicity list
# the output will be the sum calculated values of deaths agianst race ethnicity
EthnicityDeaths=getFilteredSum('Race Ethnicity','Deaths',getEthnicityList)
```

```python
# This logic calls the function getFilteredSum
# The input aprameters are Column names and year list
# the output will be the sum calculated values of deaths agianst Year
yearlyDeaths=getFilteredSum('Year','Deaths',getYearList)
```

```python
# This logic calls the function getFilteredSum
# The input aprameters are Column names and gender list
# the output will be the gender death rates
genderDeaths=getFilteredSum('Sex','Deaths',getGenderList)
```

```python
# importing matplotlib 
# This is used to plot figures
import matplotlib.pyplot as plt
```

```python
# Figure() is used to give the length and breadth of the figure
# .bar() defines the plot as bar graph
# xlabel, ylabel and title are used to name the titles fro x,y axis and the title of the graph
_=plt.figure(figsize=(15,18)) 
_=dframe.groupby(['Leading Cause'])['Deaths'].sum().plot(kind='barh')
_=plt.xlabel('Number of Deaths')
_=plt.title('Data shows the Number of deaths caused')
```

```python
# Line Graph
# Figuresize() is used to give the length and breadth of teh figure
# .plot() defines the plot as line graph
# xlabel, ylabel and title are used to name the titles fro x,y axis and the title of the graph
_=plt.figure(figsize=(10,8))
_=plt.plot(getYearList, yearlyDeaths)
_=plt.xlabel("Year")
_=plt.ylabel("Number of Deaths(#)")
_=plt.title('Data shows Number of Deaths Yearly (2007-2014)')
```

```python
# Pie Chart
# .plot() defines the plot as line graph
#title is used to name the titlesof the pie  chart
values = EthnicityDeaths
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = dframe['Race Ethnicity'].unique()
explode = [0, 0, 0, 0.0, 0.0,0]
plt.figure(figsize=(10,8))
plt.pie(values, colors=colors, labels= labels,explode=explode,counterclock=False, autopct='%1.1f%%')
plt.title('Data shows Percentage of Deaths based on Race Ethnicity')
plt.show()
```
