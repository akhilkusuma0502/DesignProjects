# Bar Graph 

```python
# Figure() is used to give the length and breadth of the figure
# .bar() defines the plot as bar graph
# xlabel, ylabel and title are used to name the titles fro x,y axis and the title of the graph
_=plt.figure(figsize=(20,10)) 
_=datasource.groupby('Year')['Total'].sum().plot.bar(fontsize=18)
_=plt.ylabel('Number of Filings made',fontsize=18)
_=plt.xlabel('Year',fontsize=18)
_=plt.title('Data Shows Yearly Filings of all business types from 1991-2019',fontsize=20)
```

# Observations 1.
In 2011, There are highest Filings happened considering all business types and all authorizations.

```python
# Figure() is used to give the length and breadth of the figure
# .bar() defines the plot as bar graph
# xlabel, ylabel and title are used to name the titles fro x,y axis and the title of the graph
_=plt.figure(figsize=(20,10)) 
_=datasource.groupby('Business Type Description')['Total'].sum().plot.barh(fontsize=18,color='coral')
_=plt.ylabel('Business Types',fontsize=22)
_=plt.xlabel('Number of Filings',fontsize=22)
_=plt.title('Data Shows Filings made by different Business Groups',fontsize=35)
```

# Observation 2: 
By the above Chart , We can observe that Business Entity Filings are larger among all other groups.

# Pie Chart to Validate Hypothesis.

# Hypothesis 1:  True
# Authority Type Domestic Filings are greater than Foreign Filings.

# Below shows quantitative evidence 

```python
# .pie() defines the plot as pie chart
# Title describes the Title of the chart.
_=plt.figure(figsize=(30,10))
colors = ['m', 'g', 'r']
plt.pie(datasource.groupby('Authority Type Description')['Total'].sum(),labels=datasource['Authority Type Description'].unique(), colors=colors,counterclock=False, autopct='%1.1f%%',textprops={'fontsize': 30})
plt.title('Pie Chart shows Authority Type Domestic Filings for all Business entities from 1991-2019',fontsize=35)
plt.show()
```

# Line Graph to validate Remaining 2 Hypothesis

# Hypothesis 2: True
June month recorded highest filings and is greater than all other monthly business certificate filings.

# Hypothesis 3: True 
November  Filings are lesser than every other month certificate filings.

# Below shows quantitative evidence 

```python
list_of_months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

I am making a list of all months which are columns in the datasource. So that i can loop over and can get particular column sum to validate hypothesis.

```python
monthlyFilings={}
for column in list_of_months:
    monthlyFilings[column]=datasource[column].sum()
```

monthlyFilings is a dictionary storing values for Each month as key and value pair.

```python
## .plot() defines the plot as pie chart
_=plt.figure(figsize=(20,10)) 
_=plt.plot(list(monthlyFilings.keys()), list(monthlyFilings.values()), color='g')
_=plt.xlabel('Months',fontsize=35)
_=plt.ylabel('Filings(#)',fontsize=35)
_=plt.title('Data Shows Monthly Business Entity Filings over all years.', fontsize=35)
_=plt.ylim(ymin=0)   ##Setting the minimum tick value to 0
_=plt.xticks(fontsize=22)
_=plt.yticks(fontsize=22)
```

# As i have observed, There is a significant increase from May to June and thus June resulted highest Number of filings and also there is trend drop over two months which are Sept-Oct, Oct-Nov and thus Nov resulted less sales Which Proves the Hypothesis 2 and 3 Presented.
